```python
import pandas as pd # impor modul Pandas untuk manipulasi data 
import matplotlib.pyplot as plt # impor modul Matplotlib untuk visualisasi data. 

# membaca file CSV menjadi DataFrame
df = pd.read_csv('data_product.csv')

# Bar chart
plt.figure(figsize=(10, 6)) # mengatur ukuran gambar 10x6 inci menggunakan modul Matplotlib.
df['Address'].value_counts().plot(kind='bar') # Menghitung jumlah kemunculan tiap nilai dalam kolom 'Address' menggunakan metode value_counts() dan membuat bar chart dengan menggambar grafik batang menggunakan metode plot()
plt.xlabel('Address') # melabeli sumbu x dengan teks 'Address'.
plt.ylabel('Count') # melabeli sumbu y dengan teks 'Count'.
plt.title('Distribution of Products by Address') # membust judul grafik dengan teks 'Distribution of Products by Address'.
plt.savefig('bar_chart.jpg') # Menyimpan chart dalam format JPG dengan nama file 'bar_chart.jpg'.
plt.show() # menampilkan chart



```

```python
import pandas as pd # impor modul Pandas untuk manipulasi data 
import matplotlib.pyplot as plt # impor modul Matplotlib untuk visualisasi data. 

# membaca file CSV menjadi DataFrame
df = pd.read_csv('data_product.csv')

# Pie chart
plt.figure(figsize=(10, 6)) # mengatur ukuran gambar 10x6 inci menggunakan modul Matplotlib
df['Rating'].value_counts().plot(kind='pie', autopct='%1.1f%%')  # Membuat pie chart dengan persentase
plt.ylabel('') # Menghapus label sumbu Y
plt.title('Distribution of Products by Rating') # Memberikan judul pada chart
plt.savefig('pie_chart.jpg') # Menyimpan pie chart dalam format JPg
plt.show() # Menampilkan chart
```
